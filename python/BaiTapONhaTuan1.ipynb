```python
#13
def tron_mang(a, b):
    c = []
    n = min(len(a), len(b))

    for i in range(n):
        c.append(a[i] + b[i])

    c.extend(a[n:])
    c.extend(b[n:])
    return c

a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]

print(tron_mang(a, b))
```

```python
#14
import random

m, n = 3, 4
a = [[random.randint(1, 10) for _ in range(n)] for _ in range(m)]

# a. Xuất ma trận
for row in a:
    print(row)

# b. dòng k
k = 1
print(a[k])

# c. cột k
print([a[i][k] for i in range(m)])

# d. dòng có tổng lớn nhất
max_row = max(a, key=sum)
print(max_row)

# e. cột tích nhỏ nhất
min_product = float('inf')
col_index = -1

for j in range(n):
    t = 1
    for i in range(m):
        t *= a[i][j]
    if t < min_product:
        min_product = t
        col_index = j

print("Cột:", col_index)

# f. dòng chẵn, cột lẻ
for i in range(m):
    for j in range(n):
        if i % 2 == 0 and j % 2 == 1:
            print(a[i][j], end=" ")

# g. TBC phần tử chẵn dòng lẻ
tong = dem = 0
for i in range(m):
    if i % 2 == 1:
        for j in range(n):
            if a[i][j] % 2 == 0:
                tong += a[i][j]
                dem += 1
print(tong/dem if dem else 0)

# h. TBC biên
tong = dem = 0
for i in range(m):
    for j in range(n):
        if i == 0 or i == m-1 or j == 0 or j == n-1:
            tong += a[i][j]
            dem += 1
print(tong/dem)

# i. không thuộc biên
tong = dem = 0
for i in range(1, m-1):
    for j in range(1, n-1):
        tong += a[i][j]
        dem += 1
print(tong/dem if dem else 0)
```

```python
#15
class SinhVien:
    def __init__(self, ma, ten, namsinh, diem):
        self.ma = ma
        self.ten = ten
        self.namsinh = namsinh
        self.diem = diem

# danh sách
ds = [
    SinhVien("02DH0001", "Lan", 2004, 6.5),
    SinhVien("01CD0002", "Phan Anh", 2003, 4.5),SinhVien("02DH0003", "Lam", 2004, 6.5),
]

# c. đủ lên lớp
print(sum(1 for sv in ds if sv.diem >= 5))

# d. đủ 20 tuổi
for sv in ds:
    if 2026 - sv.namsinh >= 20:
        print(sv.ten)

# e. hệ DH
print(sum(1 for sv in ds if sv.ma[2:4] == "DH"))

# f. tên Lan
print(sum(1 for sv in ds if "Lan" in sv.ten))

# g. họ Phan
print(sum(1 for sv in ds if sv.ten.startswith("Phan")))
```

```python
#BTVN2
# a. tam giác trên đường chéo phụ
def tong_tren_cheo_phu(a, n):
    tong = 0
    for i in range(n):
        for j in range(n):
            if i + j <= n - 1:
                tong += a[i][j]
    return tong

# e. đối xứng
def doi_xung(a, n):
    for i in range(n):
        for j in range(n):
            if a[i][j] != a[j][i]:
                return False
    return True

# f. chéo chính tăng
def cheo_chinh_tang(a, n):
    return all(a[i][i] < a[i+1][i+1] for i in range(n-1))

# h. chéo phụ giảm
def cheo_phu_giam(a, n):
    return all(a[i][n-i-1] > a[i+1][n-i-2] for i in range(n-1))


# ====== MENU ======
while True:
    print("\n===== MENU =====")
    print("1. Tổng tam giác trên chéo phụ")
    print("2. Đổi trị tuyệt đối ma trận")
    print("3. Thay số chẵn bằng 99")
    print("4. Kiểm tra toàn chẵn")
    print("5. Kiểm tra đối xứng")
    print("6. Kiểm tra chéo chính tăng")
    print("7. In tam giác dưới chéo phụ")
    print("8. Kiểm tra chéo phụ giảm")
    print("0. Thoát")

    chon = int(input("Chọn chức năng: "))

    if chon == 0:
        print("Thoát chương trình!")
        break

    else:
        n = int(input("Nhập kích thước ma trận n: "))
        a = []
        print("Nhập ma trận:")
        for i in range(n):
            row = list(map(int, input().split()))
            a.append(row)

        if chon == 1:
            print("Tổng:", tong_tren_cheo_phu(a, n))

        elif chon == 2:
            for i in range(n):
                for j in range(n):
                    if a[i][j] < 0:
                        a[i][j] = abs(a[i][j])
            print("Ma trận sau khi đổi:")
            for row in a:
                print(row)

        elif chon == 3:
            for i in range(n):
                for j in range(n):
                    if a[i][j] % 2 == 0:
                        a[i][j] = 99
            print("Ma trận sau khi thay:")
            for row in a:
                print(row)

        elif chon == 4:
            all_even = all(a[i][j] % 2 == 0 for i in range(n) for j in range(n))
            print("Toàn chẵn:", all_even)

        elif chon == 5:
            print("Đối xứng:", doi_xung(a, n))

        elif chon == 6:
            print("Chéo chính tăng:", cheo_chinh_tang(a, n))

        elif chon == 7:
            print("Tam giác dưới chéo phụ:")
            for i in range(n):
                for j in range(n):
                    if i + j >= n - 1:
                        print(a[i][j], end=" ")
                print()

        elif chon == 8:
            print("Chéo phụ giảm:", cheo_phu_giam(a, n))

        else:
            print("Lựa chọn không hợp lệ!")
```
